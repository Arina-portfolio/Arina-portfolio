## Функция для расчета t-test

```python
def test_calc(r1, r2, alpha = 0.05):
    
    s,p = ttest_ind(r1,r2)
    
    if p < alpha:
        print("Гипотеза H0 не подтверждается: Статистическая разница подтверждена")
        print("Разница средних = ", r1.mean()-r2.mean())
    else:
        print("Гипотеза H0 подтверждается: Средние статистически равны")
        print("Разница средних = 0")
        
    #return s, p
```

## Функция Манна-Уитни для сравнения распределений 

```python
def mann_whitney_func(r1, r2, alpha=0.05):
    
    s, p = mannwhitneyu(r1, r2)
    
    if p<alpha:
        print('Распределения не равны', ' p= ', p)
    else:
        print('Распределения равны', ' p= ', p)
```

## Функция Хи-квадрат для оценки статистической значимости разницы в конверсии

```python
from scipy import stats
import statsmodels.stats.proportion as proportion
```

```python
def chi_sqr(p, r, alpha=0.05):
    
    chi2stat, pval, table = proportion.proportions_chisquare(p, r)
    
    if abs(pval) < alpha:
        print()
        print('Гипотеза Н0 не подтверждена. Рост конверсии статистически значимый', 'pval = ',pval)
        
    else:
        print()
        print('Гипотеза Н0 подтверждена.Статистически значимой разницы не выявлено.', 'pval = ',pval)
```
